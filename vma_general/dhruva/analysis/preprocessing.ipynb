## Calculation of Inital Movement Vectors 

```python
import seaborn as sns 
import pandas as pd 
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt
import math
```

```python
def cart2pol(x, y):
    r = []
    theta = []
    for i in range(len(x)): 
        r.append(np.sqrt((x[i])**2 + (y[i])**2))
        theta.append(np.arctan2(y[i], x[i]))
    return(r, theta)
```

```python
movement_dfs = []
trial_dfs = []
n_backward_reaches = {}

num_participants = 20

# load in all participants movement and trial data
for i in range(num_participants):
     if i in [5, 11, 18]:
        continue 
     movement_df = pd.read_csv(f'../data/data_movements_{i}.csv')
     # count and filter out negative reaches
     backward_reaches = movement_df.loc[(movement_df.state == 'reach') & (movement_df.y < 0)].drop_duplicates('trial')
     n_backward_reaches[i] = len(backward_reaches)
     movement_df = movement_df.loc[~movement_df.trial.isin(backward_reaches.trial)].copy() # ~ means 'not'. It is removing all backward reaches.
     movement_df['participant'] = i
     movement_dfs.append(movement_df)
     trial_dfs.append(pd.read_csv(f'../data/data_trials_{i}.csv'))

for df in trial_dfs:
    df['trial'] = df.trial - 1
```

```python
n_backward_reaches
```

```python
# remove backward reaches 
movement_dfs_no_back = []
for d in movement_dfs: 
    d = d.loc[(d.state == 'reach') & (d.y >= 0)].copy()
    movement_dfs_no_back.append(d)
```

### Interpolate Data

```python
def interpolation(df): 

    t = df["time"]
    x = df["x"]
    y = df["y"]
    trial = df['trial']

    xs = CubicSpline(t, x)
    ys = CubicSpline(t, y)

    tt = np.linspace(t.min(), t.max(), 100)
    xx = xs(tt)
    yy = ys(tt)


    interpolated_df = pd.DataFrame({'time': tt, 'x': xx, 'y': yy})
 

    return interpolated_df

interpolated_dfs = []

for i, movement_df in enumerate(movement_dfs_no_back): 
    interpolated_df = movement_df.groupby('trial').apply(interpolation)
    interpolated_df['participant'] = i
    interpolated_dfs.append(interpolated_df)
    #fig, ax = plt.subplots(1, 2, squeeze=False)
    #ax[0, 0].plot(x, y, '.')
    #ax[0, 1].plot(xx, yy, '.')
    #plt.show()
    # fig, ax = plt.subplots(1, 2, squeeze=False)
    # sns.scatterplot(data=movement_df, x='x', y='y', ax=ax[0,0])
    # sns.scatterplot(data=interpolated_df, x='x', y='y', ax=ax[0,1])
    # plt.show()
    # fig, ax = plt.subplots(1, 2, squeeze=False)
    # ax[0, 0].plot(x, y, '.')
    # ax[0, 1].plot(xx, yy, '.')
    # plt.show()
```

```python
# prep the previous uncertainty information in the first trial df as uncertainty is trial matched across participants.

previous_uncertainty = ['no_uncertainty'] * 21

for i in range(21, 300):
    prev_row = trial_dfs[0].iloc[i - 1]
    if prev_row.no_uncertainty == True:
        previous_uncertainty.append('no_uncertainty')
    elif prev_row.low_uncertainty == True: 
        previous_uncertainty.append('low_uncertainty')
    elif prev_row.high_uncertainty == True: 
        previous_uncertainty.append('high_uncertainty')
    elif prev_row.unlimited_uncertainty == True: 
        previous_uncertainty.append('unlimited_uncertainty')
    else: 
        previous_uncertainty.append('unlimited_uncertainty')
    

trial_dfs[0]['previous_uncertainty'] = previous_uncertainty
```

### Calculate initial movmement vector

IMV is defined at the hand anlge at 4.5cm of the reach (or first measure there-after). This is the theta before midpoint feedback is provided and may be incorporated. This method of calculation IMV was chosen because velocity profiles of reaches seem to initiate with a curve to the left; presumably this is the result of the biomechanical intricacies resulting from the experimental apparatus. 

```python
# convert coordinate sytems. 
r, theta = cart2pol(micro_plot_df.x.values, micro_plot_df.y.values)
micro_plot_df['r'] = r
micro_plot_df['theta'] = theta
micro_plot_df['theta'] = micro_plot_df.theta.apply(math.degrees)
micro_plot_df['imv'] = abs(90 - micro_plot_df.theta)
```

```python
# change trial values to between 139 and 200 for last 60 trials of adaptation where the kinematics suggest there is stratification. 
# change trial values to between 19 and 200 for the whole adaptation block. 
imv = micro_plot_df.loc[(micro_plot_df.r >= 4.5) & (micro_plot_df.r < 5) & (micro_plot_df.trial > 139) & (micro_plot_df.trial < 200)]
imv = imv.groupby('trial').agg({'r' : 'first', 'imv' : 'first', 'previous_uncertainty' : 'first'}).reset_index()
```

```python
plt.figure(figsize=(10, 6))
# Create a color palette based on previous_uncertainty levels
palette = sns.color_palette("Set1", n_colors=len(imv['previous_uncertainty'].unique()))

# Plot the behavioral data as a line
sns.lineplot(x="trial", y="imv", data=imv, hue='previous_uncertainty', palette=palette)

# Create a scatter plot for color-coding the behavioral data points
sns.scatterplot(x="trial", y="imv", hue="previous_uncertainty", data=imv, palette=palette, legend=False)
# Labels
plt.xlabel('Trial')
plt.ylabel('Initial Movement Vector')
plt.title('Initial Movement Vector by Trial')
plt.legend(title="Previous Uncertainty", bbox_to_anchor=(0.995, 0.85), loc='center left')
plt.grid(False)
```

Here initial movement vector (IMV) is calculated by finding the point of peak velocity. The angular difference between the optimal movement vector and first point at which velocity exceeds 5% of peak velocity after a 2cm reach is considered the inital movement vector. 

Midpoint theta is defined and calculated as the first measured angle at 5cm (or higher until first measure is found).

### Calculate error experienced at midpoint
This is defined as the departure of the actual hand angle from the ideal and hangle plus the rotation applied on that trial.  

```python
midpoint = micro_plot_df.loc[(micro_plot_df.r >= 5)]
midpoint = midpoint.groupby('trial').agg({'r' : 'first', 'theta' : 'first', 'rot' : 'first'}).reset_index()
midpoint['error'] = abs(90 - midpoint.theta)
midpoint['exp_error'] = midpoint.error + midpoint.rot
```

```python
midpoint = pd.merge(midpoint, trial_dfs[0], on='trial', how='left').drop_duplicates()
```

```python
current_uncertainty = []
start_trial = 0
end_trial = 199
for i in range(start_trial, 300):
    row = trial_dfs[0].iloc[i]
    if row.no_uncertainty == True:
        current_uncertainty.append('no_uncertainty')
    elif row.low_uncertainty == True: 
        current_uncertainty.append('low_uncertainty')
    elif row.high_uncertainty == True: 
        current_uncertainty.append('high_uncertainty')
    elif row.unlimited_uncertainty == True: 
        current_uncertainty.append('unlimited_uncertainty')
    else: 
        current_uncertainty.append('baseline/washout')

midpoint['current_uncertainty'] = current_uncertainty
```

```python
midpoint['endpoint_theta'] = abs(90 - midpoint.endpoint_theta)
midpoint['fb_int'] = midpoint.endpoint_theta - midpoint.error
```
